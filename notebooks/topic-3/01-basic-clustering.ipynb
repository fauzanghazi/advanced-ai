# Basic Clustering
## DBSCAN

### Import necessary libraries

```python
from numpy import unique, where
import matplotlib.pyplot as plt
```

### Import and display the dataset

We'll use make_moons dataset, which is a built-in dataset in scikit-learn library.

Can read more about it [here](https://scikit-learn.org/stable/modules/generated/sklearn.datasets.make_moons.html).

```python
from sklearn.datasets import make_moons
X, _ = make_moons(noise=0.05, random_state=0, n_samples=1000)

plt.scatter(X[:,0], X[:,1]) # visualize the dataset
```

#### Model

```python
from sklearn.cluster import DBSCAN

model = DBSCAN(eps=0.2, min_samples=9)

yhat = model.fit_predict(X)

clusters = unique(yhat)

for cluster in clusters:
    row_ix = where(yhat == cluster)
    plt.scatter(X[row_ix, 0], X[row_ix, 1])
plt.title("DBSCAN")
plt.show()
```

#### k-Means Clustering

```python
from sklearn.cluster import KMeans

model = KMeans(n_clusters=2)

model.fit(X)

yhat = model.predict(X)

clusters = unique(yhat)

for cluster in clusters:
    row_ix = where(yhat == cluster)
    plt.scatter(X[row_ix, 0], X[row_ix, 1])
plt.title("k-means clustering")
plt.show()
```

#### Hierarchical Clustering

```python
from sklearn.cluster import AgglomerativeClustering

model = AgglomerativeClustering(n_clusters=2)
yhat = model.fit_predict(X)

clusters = unique(yhat)
for cluster in clusters:
    row_ix = where(yhat == cluster)
    plt.scatter(X[row_ix, 0], X[row_ix, 1])
plt.title("Agglomerative Clustering")
plt.show()
```

#### Gaussian Mixture Model (GMM)

```python
from sklearn.mixture import GaussianMixture

model = GaussianMixture(n_components=2)
model.fit(X)
yhat = model.predict(X)

clusters = unique(yhat)
for cluster in clusters:
    row_ix = where(yhat == cluster)
    plt.scatter(X[row_ix, 0], X[row_ix, 1])
plt.title("Gaussian Mixture Model")
plt.show()
```

#### Mean Shift Clustering

```python
from sklearn.cluster import MeanShift

model = MeanShift()
yhat = model.fit_predict(X)

clusters = unique(yhat)
for cluster in clusters:
    row_ix = where(yhat == cluster)
    plt.scatter(X[row_ix, 0], X[row_ix, 1])
plt.title("Mean Shift Clustering")
plt.show()
```
